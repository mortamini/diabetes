```python
path1 = '/content/drive/MyDrive/'
dc = os.path.join(path1,'imputated.csv')
diabetes = pd.read_csv(dc)
```

```python
diabetes = np.array(diabetes)
diabetes
```

```python
min_list = []
for i in range(0,len(diabetes)):
  miin = np.min(diabetes[i][4:7])
  min_list.append(miin)
```

```python
max_list = []
for i in range(0,len(diabetes)):
  maax = np.max(diabetes[i][4:7])
  max_list.append(maax)
```

```python
diabetes = pd.DataFrame(diabetes,columns=['date','case','Reg_ins','NPH_ins','Pr_brkfst_glu','Pr_lnch_glu','Pr_sup_glu'])
```
